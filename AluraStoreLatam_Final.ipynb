### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Ingreso_Total1=float(tienda['Precio'].sum())
Ingreso_Total2=float(tienda2['Precio'].sum())
Ingreso_Total3=float(tienda3['Precio'].sum())
Ingreso_Total4=float(tienda4['Precio'].sum())
print(f"El total de la tienda 1 es: {Ingreso_Total1}")
print(f"El total de la tienda 2 es: {Ingreso_Total2}")
print(f"El total de la tienda 3 es: {Ingreso_Total3}")
print(f"El total de la tienda 4 es: {Ingreso_Total4}")
```

# 2. Ventas por categoría

```python
print("La categoria más vendida de la tienda 1 es:")
Cat1=tienda.groupby('Categoría del Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False).head()
Cat1
```

```python
print("La categoria más vendida de la tienda 1 es:")
Cat2=tienda2.groupby('Categoría del Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False).head()
Cat2
```

```python
print("La categoria más vendida de la tienda 3 es:")
Cat3=tienda3.groupby('Categoría del Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False).head()
Cat3
```

```python
print("La categoria más vendida de la tienda 4 es:")
Cat4=tienda4.groupby('Categoría del Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False).head()
Cat4
```

# 3. Calificación promedio de la tienda


```python
prom_cal_t1=float(round(tienda['Calificación'].mean(),2))
prom_cal_t2=float(round(tienda2['Calificación'].mean(),2))
prom_cal_t3=float(round(tienda3['Calificación'].mean(),2))
prom_cal_t4=float(round(tienda4['Calificación'].mean(),2))
print(f"La calificación promedio de la tienda 1 es: {prom_cal_t1}")
print(f"La calificación promedio de la tienda 2 es: {prom_cal_t2}")
print(f"La calificación promedio de la tienda 3 es: {prom_cal_t3}")
print(f"La calificación promedio de la tienda 4 es: {prom_cal_t4}")
```

# 4. Productos más y menos vendidos

```python
print("Se muestra el producto más vendido y el menos vendido respectivamente en la tienda 1")
Tabla1=tienda.groupby('Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False)
pd.concat([Tabla1.head(1), Tabla1.tail(1)])
```

```python
print("Se muestra el producto más vendido y el menos vendido respectivamente en la tienda 2")
Tabla2=tienda2.groupby('Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False)
pd.concat([Tabla2.head(1), Tabla2.tail(1)])
```

```python
print("Se muestra el producto más vendido y el menos vendido respectivamente en la tienda 3")
Tabla3=tienda3.groupby('Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False)
pd.concat([Tabla3.head(1), Tabla3.tail(1)])
```

```python
print("Se muestra el producto más vendido y el menos vendido respectivamente en la tienda 4")
Tabla4=tienda.groupby('Producto').size().to_frame('cantidad').sort_values('cantidad',ascending=False)
pd.concat([Tabla4.head(1), Tabla4.tail(1)])
```

# 5. Envío promedio por tienda

```python
cost_envi1 =round(float(tienda["Costo de envío"].mean()),1)
cost_envi2 =round(float(tienda2["Costo de envío"].mean()),1)
cost_envi3 =round(float(tienda3["Costo de envío"].mean()),1)
cost_envi4 =round(float(tienda4["Costo de envío"].mean()),1)

print(f"El costo de envio promedio de la tienda 1 es: {cost_envi1}")
print(f"El costo de envio promedio de la tienda 2 es: {cost_envi2}")
print(f"El costo de envio promedio de la tienda 3 es: {cost_envi3}")
print(f"El costo de envio promedio de la tienda 4 es: {cost_envi4}")
```
